# Separacin Dataset Principal - Independiente (80-20%)

```python
import random
import screed
```

```python
def read_sequences_from_file(filename):
    all_sequences = []
    for record in screed.open(filename):
        sequence = record.sequence
        all_sequences.append(sequence)   
    return all_sequences
```

```python
filename_ACP = 'Databases/ACP.fasta'
filename_noACP = 'Databases/noACP.fasta'
```

```python
all_sequences_ACP = read_sequences_from_file(filename_ACP) # ACP
all_sequences_noACP = read_sequences_from_file(filename_noACP) # noACP
```

```python
random.Random(2).shuffle(all_sequences_ACP)
random.Random(2).shuffle(all_sequences_noACP)
```

```python
ACP_Main = all_sequences_ACP[:758]
ACP_Independiente = all_sequences_ACP[758:]
```

```python
noACP_Main = all_sequences_noACP[:758]
noACP_Independiente = all_sequences_noACP[758:]
```

### ACP

```python
with open('Databases/Principal/ACP_Main.txt', 'w') as filehandle:
    for listitem in ACP_Main:
        filehandle.write('%s\n' % listitem)
        
with open('Databases/Independiente/ACP_independiente.txt', 'w') as filehandle:
    for listitem in ACP_Independiente:
        filehandle.write('%s\n' % listitem)
```

### noACP

```python
with open('Databases/Principal/noACP_Main.txt', 'w') as filehandle:
    for listitem in noACP_Main:
        filehandle.write('%s\n' % listitem)
        
with open('Databases/Independiente/noACP_Independiente.txt', 'w') as filehandle:
    for listitem in noACP_Independiente:
        filehandle.write('%s\n' % listitem)
```
